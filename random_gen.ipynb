## Random instrument setup at locations

````python
import pandas as pd
import random

# --------------------------------------------------
# Configuration
# --------------------------------------------------

units = ["A1", "A2", "A3", "A4", "A5"]

instrument_classes = [
    "ALPHA",
    "BETA",
    "GAMMA",
    "DELTA"
]

# number of instruments to generate
N_INSTRUMENTS = 100

random.seed(42)

# --------------------------------------------------
# Generate instruments
# --------------------------------------------------

rows = []

for i in range(1, N_INSTRUMENTS + 1):

    instrument_id = f"I{i:03d}"

    unit_id = random.choice(units)

    instrument_class = random.choice(instrument_classes)

    rows.append({
        "instrument_id": instrument_id,
        "unit_id": unit_id,
        "instrument_class": instrument_class
    })

# --------------------------------------------------
# Create dataframe
# --------------------------------------------------

instruments_df = pd.DataFrame(rows)

# --------------------------------------------------
# Sort for readability
# --------------------------------------------------

instruments_df = instruments_df.sort_values(
    by=["unit_id", "instrument_class", "instrument_id"]
)

# --------------------------------------------------
# Save CSV
# --------------------------------------------------

instruments_df.to_csv(
    "instruments.csv",
    index=False
)

# --------------------------------------------------
# Preview
# --------------------------------------------------

print(instruments_df)
````

## N months of random production statistics 

```text

The date range etc. can be set in the code by altering:

START_DATE = date(2026, 1, 1)
END_DATE = date(2026, 4, 30)

As well as the production volumes can be set for a equipment_class:

class_volume = {
    "ALPHA": (80, 180),
    "BETA":  (40, 120),
    "GAMMA": (20, 90),
    "DELTA": (10, 60)
}

```

````python
import pandas as pd
import random
from datetime import date, timedelta

# --------------------------------------------------
# Configuration
# --------------------------------------------------

START_DATE = date(2026, 1, 1)
END_DATE = date(2026, 4, 30)

random.seed(42)

# --------------------------------------------------
# Load instruments
# --------------------------------------------------

instruments_df = pd.read_csv("instruments.csv")

# --------------------------------------------------
# Generate date range
# --------------------------------------------------

dates = pd.date_range(
    START_DATE,
    END_DATE,
    freq="D"
)

# --------------------------------------------------
# Class-based production profiles
# --------------------------------------------------

class_volume = {
    "ALPHA": (80, 180),
    "BETA":  (40, 120),
    "GAMMA": (20, 90),
    "DELTA": (10, 60)
}

# --------------------------------------------------
# Generate production rows
# --------------------------------------------------

rows = []
result_counter = 1

for _, instrument in instruments_df.iterrows():
    instrument_id = instrument["instrument_id"]
    instrument_class = instrument["instrument_class"]

    min_volume, max_volume = class_volume[instrument_class]

    for run_date in dates:
        # small chance that an instrument has no production that day
        if random.random() < 0.08:
            continue

        result_count = random.randint(min_volume, max_volume)

        rows.append({
            "result_id": f"R{result_counter:05d}",
            "run_date": run_date.strftime("%Y-%m-%d"),
            "instrument_id": instrument_id,
            "result_count": result_count
        })

        result_counter += 1

# --------------------------------------------------
# Create dataframe
# --------------------------------------------------

production_df = pd.DataFrame(rows)

# --------------------------------------------------
# Save CSV
# --------------------------------------------------

production_df.to_csv(
    "production_results.csv",
    index=False
)

# --------------------------------------------------
# Preview
# --------------------------------------------------

print(production_df.head())
print(f"Rows generated: {len(production_df)}")
````

## Create equipment.csv from instruments.csv
```text
The equipment.csv file is used for creating a user import file and clearly shows how the complexity grows with number of instruments, access rights, equipment classes. 
```

````python
import pandas as pd

# --------------------------------------------------
# Load instruments
# --------------------------------------------------

instruments_df = pd.read_csv("instruments.csv")

# --------------------------------------------------
# Create equipment table
# Unique unit_id + instrument_class combinations
# --------------------------------------------------

equipment_df = (
    instruments_df[
        ["unit_id", "instrument_class"]
    ]
    .drop_duplicates()
    .rename(columns={
        "instrument_class": "equipment_class"
    })
    .sort_values(
        by=["unit_id", "equipment_class"]
    )
    .reset_index(drop=True)
)

# --------------------------------------------------
# Save CSV
# --------------------------------------------------

equipment_df.to_csv(
    "equipment.csv",
    index=False
)

# --------------------------------------------------
# Preview
# --------------------------------------------------

print(equipment_df)
````

## Generate Users

````python
import pandas as pd
import random

# --------------------------------------------------
# Configuration
# --------------------------------------------------

N_USERS = 100

units = [
    "A1",
    "A2",
    "A3",
    "A4",
    "A5"
]

first_names = [
    "Alice",
    "Bob",
    "Charlie",
    "Diana",
    "Eric",
    "James",
    "Emma",
    "Lucas",
    "Olivia",
    "Noah",
    "Sophia",
    "Liam",
    "Mia",
    "William",
    "Ella",
    "Benjamin",
    "Ava",
    "Henry",
    "Isabella",
    "Jack",
    "Evelyn",
    "Leo",
    "Grace",
    "Oscar",
    "Chloe",
    "Alexander",
    "Emily",
    "Daniel",
    "Hannah",
    "Sebastian"
]

random.seed(42)

# --------------------------------------------------
# Generate users
# --------------------------------------------------

rows = []

for i in range(1, N_USERS + 1):

    user_id = f"U{i}"

    name = random.choice(first_names)

    employed_at = random.choice(units)

    rows.append({
        "user_id": user_id,
        "name": name,
        "employed_at": employed_at
    })

# --------------------------------------------------
# Create dataframe
# --------------------------------------------------

users_df = pd.DataFrame(rows)

# --------------------------------------------------
# Sort users
# --------------------------------------------------

users_df = users_df.sort_values(
    by=["employed_at", "user_id"]
)

# --------------------------------------------------
# Save CSV
# --------------------------------------------------

users_df.to_csv(
    "users.csv",
    index=False
)

# --------------------------------------------------
# Preview
# --------------------------------------------------

print(users_df)
````

## User Import Code

```text
User import demo code (Separate demo exists), only added to show how the complexity grow with instruments and accesss rights, making manual input a time consuming work.

With 100 users and a relatively small instrument environment, the generated import already expands to more than 1200 access rows.

The complexity does not primarily come from the code itself, but from the combinatorial growth of operational relationships between users, wards, access rules, and instrument classes.

```



````python
import sqlite3
import pandas as pd

conn = sqlite3.connect(":memory:")

pd.read_csv("users.csv").to_sql("users", conn, index=False, if_exists="replace")
pd.read_csv("access_rules.csv").to_sql("access_rules", conn, index=False, if_exists="replace")
pd.read_csv("equipment.csv").to_sql("equipment", conn, index=False, if_exists="replace")

query = """
SELECT
    u.user_id,
    u.name AS user_name,
    u.employed_at,
    ar.access_unit,
    ar.access_type,
    e.equipment_class

FROM users u

JOIN access_rules ar
    ON ar.employed_at = u.employed_at

JOIN equipment e
    ON e.unit_id = ar.access_unit

ORDER BY
    u.user_id,
    ar.access_unit,
    e.equipment_class
"""

rows = pd.read_sql_query(query, conn)

rows.to_csv(
    "user_import_export.csv",
    index=False,
    sep=";"
)

print(rows)
````
